```python
columns = ['ID', 'Page_name', 'Date', 'Device', 'Gender']
df3 = pd.read_csv('click_stream3.csv', names=columns, parse_dates=['Date'], dayfirst=False)
```

```python
df3.head()
```

```python
df3['Month'] = df3['Date'].dt.month
df3
```

```python
ex = df3.groupby(['Month', 'Page_name', 'Device','Gender']).agg('count')
ex.tail(30)
```

```python
gender = df3.loc[:,['Device', 'Gender','ID']].groupby(['Device', 'Gender']).size()
```

```python
gender
```

```python
plt.figure(figsize=(10, 5))
x = np.asarray([0, 1, 2, 3])
width = 0.15
i = -3
for name, group in df3.loc[:,['Page_name', 'Gender','ID']].groupby(['Gender']):
    plt.bar(x - i*width, group.groupby('Page_name').size().values, width, label=name)
    i += 1
plt.title('Влияние пола на переходы по страничкам')
plt.xticks(x, list(df3.loc[:,['Page_name', 'Gender','ID']].groupby('Page_name').groups.keys()))
plt.legend()
plt.show()
```

Вывод 1: Пол не влияет на переходы по страничкам (практивечки одинаковое кол-во мужчин и женщин переходят на следующий этап покупки)

```python
plt.figure(figsize=(10, 5))
x = np.asarray([0, 1, 2, 3])
width = 0.15
i = -3
for name, group in df3.loc[:,['Month', 'Gender','ID']].groupby(['Gender']):
    plt.bar(x - i*width, group.groupby('Month').size().values, width, label=name)
    i += 1
plt.title('Взаимосвзяь пола и месяца')
plt.xticks(x, list(df3.loc[:,['Month', 'Gender','ID']].groupby('Month').groups.keys()))
plt.legend()
plt.show()
```

Вывод 2: Во все месяцы мужчины и женщины примерно одинаково активны

```python
plt.figure(figsize=(10, 5))
x = np.asarray([0, 1])
width = 0.15
i = -1
for name, group in df3.loc[:,['Device', 'Gender','ID']].groupby(['Gender']):
    plt.bar(x - i*width, group.groupby('Device').size().values, width, label=name)
    i += 1
plt.title('Влияние пола на выбор устройства')
plt.xticks(x, list(df3.loc[:,['Device', 'Gender','ID']].groupby('Device').groups.keys()))
plt.legend()
plt.show()
```

Вывод 3: пол не влияет на выбор устроиства для совершения покупки
Таким образом, пол не оказывает особого влияния на воронку.
