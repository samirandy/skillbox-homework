```python
from sympy import *
from sympy.plotting import plot
from sympy.plotting import plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x, y, x1, x2 = symbols('x, y, x1, x2')
a2, a1, a0 = symbols('a2, a1, a0')
```

```python
f = x*(y - 5)**2
f
```

```python
diff(f, x) # нахожу производную по x
```

```python
fx = (y - 5)**2 # записываю производную
```

```python
plot(fx) # строю график производной
```

```python
solve(fx) # нахожу при каком значении y производная равна 0
```

```python
MSE = 1/3 * (((a2 * 560 + a1 * 1945 + a0) - 4.3)**2 + \
             ((a2 * 340 + a1 * 1495 + a0) - 4.9)**2 + \
             ((a2 * 343 + a1 * 1570 + a0) - 5.2)**2)
MSE
```

```python
diff(MSE, a2) # нахожу производную по a2
```

```python
diff(MSE, a1) # нахожу производную по a1
```

```python
MSEa1 = 3340.0*a0 + 5655300.0*a1 + 1424006.66666667*a2 - 15902.0
MSEa1
```
